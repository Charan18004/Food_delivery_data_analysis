```python
# üçΩÔ∏è Food Delivery Data Analysis

import pandas as pd
import sqlite3

# ======================
# Load datasets
# ======================
orders = pd.read_csv("orders.csv")
users = pd.read_json("users.json")

# ======================
# Load restaurants.sql
# ======================
conn = sqlite3.connect(":memory:")

with open("restaurants.sql") as f:
    conn.executescript(f.read())

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)

# ======================
# Merge (LEFT JOIN)
# ======================
df = orders.merge(users, on="user_id", how="left")
df = df.merge(restaurants, on="restaurant_id", how="left")

df.head()

# ======================
# Save final dataset
# ======================
df.to_csv("final_food_delivery_dataset.csv", index=False)

# ======================
# Q1 ‚Äî Highest revenue city (Gold)
# ======================
gold_df = df[df["membership"] == "Gold"]
print(gold_df.groupby("city")["total_amount"].sum().sort_values(ascending=False))

# ======================
# Q2 ‚Äî Highest avg cuisine value
# ======================
print(df.groupby("cuisine")["total_amount"].mean().sort_values(ascending=False))
```
